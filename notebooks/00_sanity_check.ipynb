```python
# --- Audio sanity checks ---

from src.audio_utils import (
    list_audio_files,
    get_turn_durations,
    compute_total_speech_time,
    compute_speaker_dominance,
    compute_turn_statistics,
)

audio_dir = "data/audio"  # adjust if needed

files = list_audio_files(audio_dir)
print("Audio files found:", len(files))

durations = get_turn_durations(audio_dir)

print("First 3 durations:")
for d in durations[:3]:
    print(d)

# Mock speaker assignment (for now)
# Later this should come from aligned metadata
mock_turns = []
speakers = ["A", "B"]

for i, d in enumerate(durations):
    mock_turns.append({
        "speaker": speakers[i % 2],
        "duration": d["duration"]
    })

total_time = compute_total_speech_time(mock_turns)
dominance = compute_speaker_dominance(mock_turns)
stats = compute_turn_statistics(mock_turns)

print("\nTotal speech time:", total_time)
print("Speaker dominance:", dominance)
print("Turn statistics:", stats)
```

```python
from src.alignment_utils import align_turns_with_audio

turns = [{"speaker": t["speaker"], "text": t["text"]} for t in d["turns"]]
aligned_turns = align_turns_with_audio(turns, "data/audio")

for turn in aligned_turns[:5]:
    print(turn)
```

```python
# --- Turn duration analysis ---

if aligned_turns:
    durations = [turn["duration"] for turn in aligned_turns if turn["duration"] is not None]

    print("Number of aligned turns:", len(aligned_turns))
    print("Min duration:", min(durations))
    print("Max duration:", max(durations))
    print("Average duration:", sum(durations) / len(durations))

    by_speaker = defaultdict(list)
    for turn in aligned_turns:
        if turn["duration"] is not None:
            by_speaker[turn["speaker"]].append(turn["duration"])

    print("\nAverage turn duration by speaker:")
    for speaker, vals in by_speaker.items():
        print(f"{speaker}: {sum(vals) / len(vals):.2f}s")

    shortest_turn = min(
        (turn for turn in aligned_turns if turn["duration"] is not None),
        key=lambda x: x["duration"]
    )
    longest_turn = max(
        (turn for turn in aligned_turns if turn["duration"] is not None),
        key=lambda x: x["duration"]
    )

    print("\nShortest turn:")
    print(shortest_turn)

    print("\nLongest turn:")
    print(longest_turn)
else:
    print("No aligned turns found.")
```
