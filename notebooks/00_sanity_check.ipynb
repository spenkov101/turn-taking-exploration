```python
from src.data_loader import load_local_dialogues
from src.turn_utils import turn_last_tokens

dialogs = load_local_dialogues()

d = dialogs[0]
print("Dialogue ID:", d["dialogue_id"])

for i, turn in enumerate(d["turns"]):
    last = turn_last_tokens([turn["text"]])[0]
    print(f'{i} [{turn["speaker"]}]: {turn["text"]} | last_token={last}')
```

```python
from src.turn_utils import speaker_balance

# convert turns to simple format
turns = [{"speaker": t["speaker"], "text": t["text"]} for t in d["turns"]]

speaker_balance(turns)
```

```python
balance = speaker_balance(turns)

print("Speaker word counts:", balance["speaker_words"])
print("Balance ratio:", balance["balance_ratio"])
```

```python
# --- Audio sanity checks ---

from src.audio_utils import (
    list_audio_files,
    get_turn_durations,
    compute_total_speech_time,
    compute_speaker_dominance,
    compute_turn_statistics,
)

audio_dir = "data/audio"  # adjust if needed

files = list_audio_files(audio_dir)
print("Audio files found:", len(files))

durations = get_turn_durations(audio_dir)

print("First 3 durations:")
for d in durations[:3]:
    print(d)

# Mock speaker assignment (for now)
# Later this should come from aligned metadata
mock_turns = []
speakers = ["A", "B"]

for i, d in enumerate(durations):
    mock_turns.append({
        "speaker": speakers[i % 2],
        "duration": d["duration"]
    })

total_time = compute_total_speech_time(mock_turns)
dominance = compute_speaker_dominance(mock_turns)
stats = compute_turn_statistics(mock_turns)

print("\nTotal speech time:", total_time)
print("Speaker dominance:", dominance)
print("Turn statistics:", stats)
```
